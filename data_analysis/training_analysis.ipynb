```python
import pandas as pd
```

```python
SESSION = "3_7__20_29_10"
folder_path = f"./training_data/{SESSION}"
EVAL = "random"

loss_df = pd.read_csv(f"{folder_path}/loss_out.csv", names=["eps", "loss"], header=None)
reward_df = pd.read_csv(f"{folder_path}/reward_out.csv", names=["eps", "reward"], header=None)
# reward_df = reward_df.iloc[8335:]
eval_df = pd.read_csv(f"{folder_path}/{EVAL}/eval_reward_out.csv", names=["eps", "reward"], header=None)
scores_df = pd.read_csv(f"{folder_path}/{EVAL}/scores.csv", names=["eps", "agent_score", "opp_score"], header=None)

loss_df.head()
reward_df.head()
eval_df.head()
print(reward_df['reward'].mean())
print(eval_df['reward'].mean())
print(scores_df['agent_score'].mean())
print(scores_df['opp_score'].mean())

```

```python
loss_df.plot(x="eps", y="loss")
reward_df.plot(x="eps", y="reward")
```

```python
loss_df["CMA"] = loss_df["loss"].expanding().mean()
reward_df["CMA"] = reward_df["reward"].expanding().mean()
```

```python
loss_df.plot(x="eps", y="CMA", xlabel="iteration", ylabel="loss", title="Mean Squared Error CMA", figsize=(10,5))
```

```python
reward_df.plot(x="eps", y="CMA", xlabel="iteration", ylabel="reward", title="Reward CMA", figsize=(10,5), ylim=(-22,-19), xlim=(0, 1000000))
```
